## Hashtag

```python
import re
```

```python
# Find out the hasktag
ht=re.findall(r"#(\w+)",text)
ht
```

```python
def hashtag_extract(text_corpus):
    hashtag=[]
    for text in text_corpus:
        ht=re.findall(r"#(\w+)",text)
        hashtag.append(ht)
        
    return hashtag    
```

```python
import nltk 
```

```python
def hashtag_freq(hashtag):
    a=nltk.FreqDist(hashtag)
    d=pd.DataFrame({'Hashtag':list(a.keys()),
                     'Freq':list(a.values())}
                     )
    d=d.nlargest(columns="Freq",n=30)
    
    return d
```

```python
hashtags=hashtag_extract(ukraine_war['content'])
```

```python
hashtags
```

```python
hashtags=sum(hashtags,[])
hashtags
```

```python
hasht=hashtag_freq(hashtags)
hasht
```

```python
# Barplot
plt.figure(figsize=(25,15))
sns.barplot(x='Hashtag',y='Freq',data=hasht)
plt.xticks(rotation=90)
plt.show()
```

```python
def clean(text):
  text=str(text).lower()
  text=re.sub('https?://\S+|www\.\S+','',text)
  text=re.sub('[%s]'%re.escape(string.punctuation),'',text)
  text=re.sub('\n|\t','',text)
  text=re.sub('\d','',text)
  text=[word for word in text.split(' ') if word not in stopword]
  text=" ".join(text)
  text=[stemmer.stem(word) for word in text.split(' ')]
  text=" ".join(text)

  return text
```

```python
ukraine_war['clear_tweet']=ukraine_war['content'].apply(clean)
```
